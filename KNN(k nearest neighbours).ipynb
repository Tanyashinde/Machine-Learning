```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
plt.style.use('fivethirtyeight')
```

```python
X,y= make_blobs(n_samples=300,centers=3)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
X_new= [5,7]
plt.scatter(X[:,0], X[:,1], c=y)
plt.scatter(X_new[0],X_new[1],c='red')
plt.show()
```

```python
from math import sqrt
from collections import Counter

dists={}

for i in range(300):
    point= X[i]
    eucl_dist= sqrt( (point[0]-X_new[0])**2 + (point[1]-X_new[1])**2)
    dists[eucl_dist]=y[i]

knn =[]
k =0

for key in sorted(dists):
    if k<5 :
        knn.append(dists[key])
        k+=1
    else :
        break

closest_dists = Counter(knn).most_common()
closest_dists
```

```python
confidence= closest_dists[0][1]* 100 /k
confidence
```
