```python
import pandas as pd
import numpy as nm
```

```python
iris= pd.read_csv("IRIS.csv")
iris
```

```python
import matplotlib.pyplot as plt
```

```python
X= iris[['sepal_length']]
y=[]
z= nm.array(iris[['species']])

for i in range(150):
     if (z[i][0]=='Iris-versicolor') :
         y.append(0)
     elif (z[i][0]=='Iris-setosa') :
         y.append(1)
     else :
         y.append(2)
y
```

```python
plt.scatter(X,y,c=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_new= iris[['sepal_length','sepal_width','petal_length','petal_width']]
x_train,x_test,y_train,y_test= train_test_split(x_new,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model= KNeighborsClassifier()
model.fit(x_train,y_train)
```

```python
y_predict =  model.predict(x_test)
y_predict
```

```python
model.score(x_test,y_test)
```
