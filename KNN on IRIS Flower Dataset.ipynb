```python
import pandas as pd
import numpy as nm
```

```python
iris= pd.read_csv("IRIS.csv")
iris
```

```python
import matplotlib.pyplot as plt
```

```python
X= iris[['sepal_length']]
y=[]
z= nm.array(iris[['species']])

for i in range(150):
     if (z[i][0]=='Iris-versicolor') :
         y.append(0)
     elif (z[i][0]=='Iris-setosa') :
         y.append(1)
     else :
         y.append(2)
y
```
